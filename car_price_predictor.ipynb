## Cleaning of Data


```python
backup_car = car.copy() # if the data got corrupted during cleaning
```

```python
car = car[car['year'].str.isnumeric()] # we will keep only those that are numeric and discard the rest of it
```

```python
# next is change datatype to int
car['year'] = car['year'].astype(int) # the datatype is changed to int
```

```python
car = car[car['Price'] != 'Ask For Price'] # Removed those rows where price is ask for price
```

```python
car['Price'] # we have to remove , and convert it into int
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car['Price'].head()
```

```python
# remove kms from kms_driven field and replacing 0
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'] != 'Petrol'] # removing inconsistent values
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car.info(
)
```

```python
car = car[~car['fuel_type'].isna()] # removed those rowns where fuel_type is null
```

```python
car.info()
```

```python
# keep first 3 words of name column
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car = car.reset_index(drop = True)
```

```python
car.describe()
```

```python
# we have seen that 75% of data is under 5 lacs and ther is one of 85 lac so it could be an outlier let's remove it
car = car[car['Price'] < 5e6].reset_index(drop = True)
```

##**Model Building**  

```python
X = car.drop(columns='Price')
```

```python
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Any ML model requires numeric features only
# So applying one hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
column_trans
```

```python
# Using linear Regression Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
# Making a pipeline
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred) # R2 score is measure of performance in regression problems
```

```python
# As the dataset is small we need to know which random state value gives us the max R2 score
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
